```python
import matplotlib.pyplot as plt
labels=['Apple','Banana','Cherry','Mango']
size=[30,25,20,25]
plt.pie(size,labels=labels,autopct='%1.1f%%')
plt.title('Fruit Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels=['Chrome','Firefox','Edge','Safari']
size=[65,20,10,5]
plt.pie(size,labels=labels,autopct='%1.1f%%')
plt.title('Browser usage shave')
plt.show()
```
